### <span style="color:orange">Importing Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

### <span style="color:orange">Making the Data and Plotting it</span>

```python
np.random.seed(0)

points = 500
points_x = np.linspace(-3, 3, points)  # // 500 values equally spaced between -3 and 3
points_y = np.sin(points_x)
```

```python
plt.scatter(x=points_x, y=points_y)
```

****We will add some noise to the data, so that each point can deviate from it's standard shape**

```python
points_y = np.sin(points_x) + np.random.uniform(-0.5, 0.5, points)

plt.scatter(points_x, points_y)
```

### <span style="color:orange">Model</span>

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
nn_model = Sequential()

nn_model.add(Dense(input_dim=1, units=50, activation='sigmoid') )
nn_model.add(Dense(units=30, activation='sigmoid') )
nn_model.add(Dense(units=1) )

nn_model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['accuracy'])
```

```python
nn_model.fit(x=points_x, y=points_y, epochs=50)
```

### <span style="color:orange">Plotting the Predictions</span>

```python
predictions = nn_model.predict(points_x)

plt.scatter(x=points_x, y=points_y)
plt.plot(points_x, predictions, 'ro')  # // to show the predictions as red circles for each prediction.
plt.show()
```
